```python
import numpy as np
import math
from numpy.linalg import inv
A = np.zeros((32,32))
for i in range(32):
    for j in range(32):
        x = np.random.uniform(-1,1)
        A[i][j] = x
B = inv(A)
print("A")
print(A)
print()
print("B")
print(B)
print()
print("AB")
print(np.dot(A,B))
print()
```
