```python
import numpy as np
from numpy.polynomial import polynomial as P
import matplotlib.pyplot as plt
#Problem 2 -- Linear interpolation
#24.2446,-145.262, 159.995, 109.283, 6967.62
def f(x):
    return 1/(1+25*x**2)
xleft = [-1.00,-0.75,-0.5,-0.25,0.0]
xright = [0.0,0.25,0.50,0.75,1.00]
yleft = [0] * len(xleft)
yright = [0] * len(xright)
#linear interpolation of xleft
#step 1 find the y values
for i in range(len(xleft)):
    yleft[i] = f(xleft[i])
matrixA = np.array([[1,-1,1,-1,1],
            [1,-0.75,.5625,-0.42187,.31640625],
            [1,-0.5,.25,-.125,.0625],
            [1,-.25,.0625,-0.015625,0.00390625],
             
             [1,0,0,0,0]])
matrixB = np.array(yleft)
a = np.linalg.solve(matrixA,matrixB)

print("Problem 1: Linear interpolation of left side")
for i in range(len(a)):
    print("a",i,"  =", "%11s" % round(a[i],5))
print()
#problem 2 curve fitting of 2nd order polynomial
for i in range(len(xright)):
    yright[i] = f(xright[i])
b = P.polyfit(xright,yright,2)
print("Problem 2: Curve fitting of right side")
for i in range(len(b)):
    print("b",i,"  =", "%10s" % round(b[i],5))
print()
    #problem 3 -- plotting
a0   =         1.0
a1   =     3.42905
a2   =     4.56314
a3   =     2.51763
a4   =       0.422
b0   =    0.96805
b1   =   -2.43526
b2   =    1.53649
def poly4th(x):
    return a0 + a1 * x + a2 * x**2 + a3* x**3 + a4* x**4
def poly2nd(x):
    return b0 + b1 * x + b2 * x**2
    #runge function
xfull = np.array(xleft + xright)
x = np.arange(-1,1.01,0.01)
fcurve, = plt.plot(x, f(x), label = "Runge function")

#interpolation
xleft = np.array([-1.00,-0.75,-0.5,-0.25,0.0])
icurve, = plt.plot(xleft, poly4th(xleft), label = "Interpolation curve")
#right
xright = np.array([0.0,0.25,0.50,0.75,1.00])
yright = np.array(yright)
bcurve, = plt.plot(xright,poly2nd(xright), label = "Best fit curve")

plt.legend(handles=[fcurve,icurve,bcurve])
plt.title("Problem 3: Runge, Interpolation, and Best Fit Curves")
plt.show()
```
