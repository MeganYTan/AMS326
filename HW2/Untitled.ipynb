```python
import numpy as np
import matplotlib.pyplot as plt
import pandas as pd
path = "C:/Users/mytan/Desktop/SBU/AMS315/HW/Question II/Book1.csv"
df = pd.read_csv(path)
```

```python
x = df['X']
y = df['Y']
plt.plot(x,y)
plt.show()
plt.scatter(x,y)
```

```python
from numpy.polynomial import polynomial as P
fit = P.polyfit(x,y,2)
print("Curve fit")
for i in range(len(fit)):
    print("b",i,"  =", "%10s" % round(fit[i],5))
```

```python
import matplotlib.pyplot as plt
b0 = fit[0]
b1 = fit[1]
b2 = fit[2]

x = np.arange(0,2500,.1)
plt.plot(x, b0 + b1*x + b2 * x**2,'o')

plt.show()
```
