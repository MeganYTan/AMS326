```python
import numpy as np
import matplotlib.pyplot as plt
import math
import random
import pandas as pd
def P(x,sd,mean):
    num1 = math.sqrt(2*math.pi*sd**2)
    num2 = (x-mean)**2
    num3 = num2/(2*mean**2)
    num4 = math.exp(-num3)
    return (1/num1) * num4
#q1
y0 = 10000000 #10 mil -- 7 zeros
yf = 1000000000000 # 12 zeros
ttot = 3380 #13 years
#mean1 = .1949/100
sd1 = .1473/100
#1 -- find mean
#2 plot each year

#PROBLEM 1 -- mean = ?
y = y0
ys1 = list()
#300000
mean1 = .3399/100
for i in range(ttot):
    y1 = y+(y*random.gauss(mean1,sd1))
    ys1.append(y1)
    y = y1

print(ys1[-1])
print(ys1[-1]>=yf)
#1 Find mean
print("Question 1")
print("Mean is %.4f%%" %(mean1*100))
#10^7 * (1+mean)^3380 = 10^11
```

```python
#2 plot Ys over t
years = np.arange(2003,2016,1)
yearvals = ys1[0::260]
plt.plot(years,yearvals)
print("Question 2")

plt.title('Graph of end of year value against year')
plt.xlabel("Year")
plt.ylabel("Company Value")
plt.show()
```
