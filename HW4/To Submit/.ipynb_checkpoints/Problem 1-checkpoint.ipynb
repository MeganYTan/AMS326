```python
import numpy as np
import matplotlib.pyplot as plt
import math
import random
import pandas as pd
def P(x,sd,mean):
    num1 = math.sqrt(2*math.pi*sd**2)
    num2 = (x-mean)**2
    num3 = num2/(2*mean**2)
    num4 = math.exp(-num3)
    return (1/num1) * num4
#q1
mean1 = .1949/100
sd1 = .2018/100
y0 = 1000000
tf = 260
x0 = 0
#P(0,sd,mean)
y = y0
ys1 = list()
for i in range(tf):
    y1 = y+y*random.gauss(mean1,sd1)
    if (y1>y):
        gain = y1-y
        y1 = y1 - (3.333/100)*gain
    ys1.append(y1)
    y = y1
    
#q2
mean2 = -.1949/100
sd2 = .6464/100
x0 = 0
#P(0,sd,mean)
y = y0
ys2 = list()
for i in range(tf):
    y2 = y+y*random.gauss(mean2,sd2)
    if(y2>y):
        gain = y2 - y
        y2 = y2 - (3.333/100)*gain
    ys2.append(y2)
    y = y2
```

```python
ts = np.arange(0,tf,1)
#plt.plot(ts,ys)
pos, = plt.plot(ts,ys1, label = 'Mean = .1949%')
neg, = plt.plot(ts,ys2, label = 'Mean = -.1949%')
plt.legend(handles = [pos,neg])
plt.xlabel('Time in days')
plt.ylabel('Portfolio value')
plt.title('Value of the portfolio as function of time in days')

plt.show()
```

```python
print("Portfolio value after 260 days with mean %f is %.2f." %(mean1,ys1[-1]))
print("Portfolio value after 260 days with mean %f is %.2f." %(mean2, ys2[-1]))
```
