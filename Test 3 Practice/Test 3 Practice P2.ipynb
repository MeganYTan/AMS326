```python
#backward Euler
import numpy as np
import matplotlib.pyplot as plt
import math
a = 100
w = 30
v = 100
k = w/v

def yprime(x,y):
    num = 1 + (y/x)**2
    return (y/x) - (k * math.sqrt(num)) #this is the new Y
#Euler = new y = y + h*yprime
ys = list()
xs = list()
y = 0
x=a
ys.append(y)
xs.append(x)
h = 0.01
while(x>0):
#for i in range(0,10): 
    #new y
    y = y - h * yprime(x,y)
    #new x?
    x = x-h 
    ys.append(y)
    xs.append(x)
plt.plot(xs,ys)
#print(ys)
#len(xs)
print(ys[1])
```

```python
#system of equation
import numpy as np
import scipy as sp
import matplotlib.pyplot as plt
import math
#define constants
a = 100
w = 30
v = 100
k = w/v
h = 0.01
def xprime(x,y):
    return -v * x/(math.sqrt(x**2+y**2))
def yprime(x,y):
    return (-v * y/math.sqrt(x**2+y**2))+w

def findNext(x,y,h):
    xnew = x +h * xprime(x,y)
    ynew = y + h * yprime(x,y)
    return xnew,ynew
#q 1
x = a
y = 0
xs = list()
ys = list()
xs.append(x)
ys.append(y)
while(x>0):
    x,y = findNext(x,y,h)
    xs.append(x)
    ys.append(y)

plt.plot(xs,ys)
plt.xlabel('x')
plt.ylabel('y')
plt.title("Trajectory of airplane")
plt.show()
```

```python
from numpy.polynomial import polynomial as P
fit = P.polyfit(xs[::-1],ys[::-1],2)
print("Curve fit")
for i in range(len(fit)):
    print("b",i,"  =", "%10s" % round(fit[i],5))
xaxis = np.arange(0,100,0.1)
def f(b0,b1,b2,x):
    return b0 + b1*x + b2* x**2
plt.plot(xaxis, f(fit[0], fit[1],fit[2],xaxis))
```

```python
yinter = np.interp(xs[::-1],xs[::-1],ys[::-1])
plt.plot(xs[::-1],yinter)
print(yinter)
```
